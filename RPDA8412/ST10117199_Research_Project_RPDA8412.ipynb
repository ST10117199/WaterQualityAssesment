# Introduction 



Water is one the most important natural resources needed for all of mankind and its rate of use has been steadily increasing though the past centuries with  this trend showing no signs of changes (The Editors of Encyclopaedia Britannica, May 28, 2023).The decreasing trend in the quantity of water available is not the only concern, now we also dealing with a growing trend in the degradation of water quality around the world. There are numerous factors that contribute to the water quality degradation, that can range from political, geographic, environmental and management factor all affect the quantity and quality of water available to use causing an overall drop in water resources available in the past decades. (Banjo A. Aderemi, Thomas O. Olwal, Julius M. Ndambuki, Sophia S. Rwanga,2023) (Gissella Bejarano, Mayank Jain, Arti Ramesh, Anand Seetharam, Aditya Mishra, June 2018).
Water come form 3 sources namely surface water, ground water and from recycled water (CDC, April 6, 2022). Ground water consist of any water body located below the surface of the earth and includes waters sources like aquafers, and wells, surface water is however water that is sourced from the surface, consisting of rivers lakes stems dams or ocean while recycled water is water from any source that has been treated to a standard for later use. 
Our water resources have a wide range of uses, and is used for irrigation, industry, mining, agricultural and household use (Vedantu, 11th Jun 2023). These big water consumers are often also some of the biggest contributors to the degradation of our water resources. This degradation can take the form of industrial waist, agricultural fertilizers and chemicals, livestock waist and mining wastewater. Other causes include deforestation, untreated sewage, hazardous substances, medical waist, global warming, Rubbish and the disposal of faecal matter and marine time traffic (SDWF,2023), (Iberdrola)
Rivers, dams and streams are particularly vulnerable to such pollution and are believed to receive more contaminants due to their ease of access (Muhammad Izhar Shah, Wesam Salah Alaloul, Abdulaziz Alqahtani, Ali Aldrees, Muhammad Ali Musarat, Muhammad Faisal Javed, 6 July 2021)
This lack of quality drinking ware is a major hazard to people health, the environment, food availability and the economy (Muhammad Izhar Shah, Wesam Salah Alaloul, Abdulaziz Alqahtani, Ali Aldrees, Muhammad Ali Musarat, Muhammad Faisal Javed, 6 July 2021) (Khanyi Mlaba, 1 February2022)

This study hopes to implement descriptive analytics to analyse the water quality trends and determine whether water quality degradation has occurred over the years. This will make use of several different Machine learning models to assess the water quality trends of available data sets in south Africa in the Mkomazi River. There are several factors that can be uses to assess the water for pollutants including Dissolved oxygen, temperature, nitrates or Phosphorus, pH scale and conductivity (SINAY, 11 May 2021) (Queensland government,7 July 2022) (AtlasScientific, 22 September 2021) (Muhammad Izhar Shah, Wesam Salah Alaloul, Abdulaziz Alqahtani, Ali Aldrees, Muhammad Ali Musarat, Muhammad Faisal Javed, 6 July 2021) (Quench,22 November 2022)
Given that most data set available often don’t test for every factor that can be used to test water quality. The models commonly built are usually determined by the indicators present in available data sets as opposed to a set parameter. This study will test and analyse the performances of the most suitable ML models discovered in the literature review to assessing the trends of water quality indictors Potassium, Total Alkalinity and the pH in a South African river.

### Random forrest 

#### PH(pH_Diss_Water)

```python
PHx = PHData.drop(['pH_Diss_Water'], axis=1)
PHy = PHData['pH_Diss_Water']
```

```python
def RFReg(testsize):
    #pefroming train test split
    x_train, x_test, y_train, y_test = train_test_split(PHx, PHy, test_size =testsize, random_state = 1)
    # Scaling the data .This used to standardize the data of the independent variables range
    s_x_train = scaler.fit_transform(x_train)
    s_x_test = scaler.transform(x_test)
    # calling the SKlearn function for Random Forest Regressor Models
    RF = RandomForestRegressor()
    # Below is used to aply the test data against the training data 
    RF.fit(s_x_train, y_train)
    #below is used to make a prediction on the test data 
    pred = RF.predict(s_x_train)
    # Below is used to get the accurcy score and return a range of answers.
    ACC = r2_score(y_train, pred)
    MAE = mean_absolute_error(y_train, pred)
    RMSE = np.sqrt(mean_absolute_error(y_train,pred))

    # storing data to be returned in a list format.
    disp=[[testsize,ACC,MAE,RMSE]]
    # returns the stored values.
    return disp
```

```python
# Define the initial testsize to 0
testsize = 0
# Define a empty list
answer=[[]]
# Create a while loop that will loop till the test size <0.95
while testsize < 0.95:
    #increments the tests size till the test size reaches 0.95 
    testsize = testsize+ 0.05
    #adds all the results of each icrement in test size to a list 
    answer = answer+RFReg(testsize)
#converts the list to a dataframe
RF_Acc= pd.DataFrame(answer)
#This just removes the first null value as this fromat creates null value by default that is not needed.
RF_Acc = RF_Acc.iloc[1:]
# Rename the columns to somthing more appropriate.
RF_Acc.rename(columns=
    {
    0:"test size",
    1:"ACC",
    2:"MAE",
    3:"RMSE"
    }
    ,inplace=True)
# Display results
RF_Acc.head(20)
```
